```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
# Basic info
df.info()
```

```python
# Summary statistics
df.describe().T
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Price')
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm', linewidths=0.2)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Example: price by type_area
sns.boxplot(x='type_area', y='price', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot price vs area
sns.scatterplot(x='value_area_sqft', y='price_range', hue='bhk', data=df)
plt.title('Price vs Area by BHK')
plt.show()
```

```python
sns.boxplot(x=df['price'])
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df['price_log'] = np.log1p(df['price'])  # log1p handles zero safely

sns.boxplot(x='type_area', y='price_log', data=df)
plt.xticks(rotation=45)
plt.ylabel('Log Price')
plt.show()
```

```python
bins = [0, 50_00_000, 1_00_00_000, 2_00_00_000, 5_00_00_000, 10_00_00_000]
labels = ['<50L', '50L-1Cr', '1-2Cr', '2-5Cr', '>5Cr']
df['price_range'] = pd.cut(df['price'], bins=bins, labels=labels)

sns.boxplot(x='type_area', y='price', data=df, hue='price_range')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x=df['price_range'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10,5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Price Distribution (Raw)')
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(np.log1p(df['price']), bins=50, kde=True)
plt.title('Price Distribution (Log-transformed)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='bhk', y=np.log1p(df['price']), data=df)
plt.title('Price vs BHK (log scale)')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='floor_num', y=np.log1p(df['price']), data=df)
plt.title('Price vs Floor Number')
plt.show()

plt.figure(figsize=(12,5))
sns.boxplot(x='total_floors', y=np.log1p(df['price']), data=df)
plt.title('Price vs Total Floors')
plt.show()
```

```python
facing_cols = [c for c in df.columns if c.startswith('facing_')]
facing_sum = df[facing_cols].sum().sort_values(ascending=False)
sns.barplot(x=facing_sum.index, y=facing_sum.values)
plt.xticks(rotation=45)
plt.title('Count of Facing Directions')
plt.show()
```

```python
extra_cols = ['extra_Garden/Park', 'extra_Pool', 'extra_Main Road']
extra_sum = df[extra_cols].sum()
sns.barplot(x=extra_sum.index, y=extra_sum.values)
plt.title('Extra Features Count')
plt.show()
```

```python
cat_cols = ['type_area', 'locality', 'floor_type']
for col in cat_cols:
    plt.figure(figsize=(12,5))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set general style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12,5)

# 1. Price distribution
fig, axes = plt.subplots(1,2, figsize=(16,5))
sns.histplot(df['price'], bins=50, kde=True, ax=axes[0])
axes[0].set_title('Price Distribution (Raw)')
sns.histplot(np.log1p(df['price']), bins=50, kde=True, ax=axes[1])
axes[1].set_title('Price Distribution (Log-transformed)')
plt.show()

# 2. Price vs BHK
sns.boxplot(x='bhk', y=np.log1p(df['price']), data=df)
plt.title('Price vs BHK (log scale)')
plt.show()

# 3. Price per sqft vs Area
sns.scatterplot(x='value_area_sqft', y='price_sqft', hue='bhk', data=df)
plt.title('Price per sqft vs Area')
plt.show()

# 4. Price vs Floor
fig, axes = plt.subplots(1,2, figsize=(16,5))
sns.boxplot(x='floor_num', y=np.log1p(df['price']), data=df, ax=axes[0])
axes[0].set_title('Price vs Floor Number')
sns.boxplot(x='total_floors', y=np.log1p(df['price']), data=df, ax=axes[1])
axes[1].set_title('Price vs Total Floors')
plt.show()

# 5. Facing directions count
facing_cols = [c for c in df.columns if c.startswith('facing_')]
facing_sum = df[facing_cols].sum().sort_values(ascending=False)
sns.barplot(x=facing_sum.index, y=facing_sum.values)
plt.xticks(rotation=45)
plt.title('Count of Facing Directions')
plt.show()

# 6. Extra features
extra_cols = ['extra_Garden/Park', 'extra_Pool', 'extra_Main Road']
extra_sum = df[extra_cols].sum()
sns.barplot(x=extra_sum.index, y=extra_sum.values)
plt.title('Extra Features Count')
plt.show()

# 7. Categorical features distribution
cat_cols = ['type_area', 'locality', 'floor_type']
for col in cat_cols:
    plt.figure(figsize=(12,5))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f'{col} Distribution')
    plt.show()


```
